Вычислительный эксперимент:

```python
import time # Получение текущего времени
import numpy as np # Работа с числовыми массивами
import matplotlib.pyplot as plt # Построение графиков
from sklearn import linear_model # Машинное обучение
reg = linear_model.LinearRegression() # Инструмент для построения модели линейной регрессии
np.random.seed(19130162) #Номер зачетки начальное состояние генератора
A = [] # Оценки свободного члена 
B = [] # Оценки кэфа регрессии
T = [] # Длительность вычислений
N = 10000 # Объем выборки
n = 1000 # число прогонов
```

```python
for i in range(n):
    t = time.time() # Измеряем время начала расчетов
    x = np.random.uniform(low=150, high=200, size=N) # Генерируем массив значений факторного признака
    e = np.random.normal(loc=0, scale=10, size=N) # Генерируем массив значений случайных отклонений от идеальной зависимости
    y = - 100 + x + e
    reg.fit(x.reshape(-1,1), y.reshape(-1,1))
    A.append(reg.intercept_[0])
    B.append(reg.coef_[0][0])
    T.append(time.time() - t)
    
```

Анализ распределения:

```python
plt.figure(figsize=(10, 6))
plt.hist(B, bins = 20, edgecolor = 'black', color = 'white')
plt.title('Histogram of B estimates')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(A, bins = 20, edgecolor = 'black', color = 'white')
plt.title('Histogram of A estimates')
plt.show()
```
